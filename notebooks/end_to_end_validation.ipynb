# End-to-End BI Data Validation Project
Simulazione CRM + ERP → Staging → Report finale  
Tecnologie: SQL + Python + Jupyter

```python
import pandas as pd
import sqlite3

# Load source data
customers = pd.read_csv('data/source_customers.csv')
orders = pd.read_csv('data/source_orders.csv')

# Create SQLite DB with 3 layers
conn = sqlite3.connect(':memory:')
customers.to_sql('source_customers', conn, index=False)
orders.to_sql('source_orders', conn, index=False)

print("✅ Layer Source caricati")
```

```python
# Source → Staging
with open('sql/source_to_staging.sql', 'r') as f:
    staging_query = f.read()
staging = pd.read_sql_query(staging_query, conn)
staging.to_sql('staging_orders', conn, index=False)

# Staging → Final Report
with open('sql/staging_to_report.sql', 'r') as f:
    report_query = f.read()
report = pd.read_sql_query(report_query, conn)
report.to_csv('reports/final_report_example.csv', index=False)

print("✅ Layer Staging e Report creati")
display(report)
```

```python
with open('sql/validation_queries.sql', 'r') as f:
    queries = f.read().split(';')

for i, q in enumerate(queries):
    if q.strip():
        result = pd.read_sql_query(q, conn)
        print(f"Query {i+1} result:")
        display(result)
```
